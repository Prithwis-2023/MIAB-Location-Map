```python
import matplotlib.pyplot as plt
import numpy as np
from astropy import units as u
from astropy import constants as const
import pandas as pd
```

```python
plt.figure(figsize=(8,6))
plt.xlabel('Proper motion velocity [km/s]',fontsize=20)
plt.ylabel('Time resolution for 1 ly GC map [yr]',fontsize=20)
plt.yscale('log')
plt.grid(True)
plt.tick_params(labelsize=15)

ly=365*24*60*60*u.s*const.c
vv = np.arange(0,401,1)*1000*u.m/u.s
res = ly/vv/(365*24*60*60*u.s)*u.yr

plt.plot(vv/(1000*u.m/u.s),1*res, label='1 ly')
plt.plot(vv/(1000*u.m/u.s),5*res, label='5 ly')
plt.plot(vv/(1000*u.m/u.s),10*res, label='10 ly')
plt.plot(vv/(1000*u.m/u.s),100*res, label='100 ly')

plt.legend(loc=1,prop = {'size':12})
plt.savefig('./GC time resolution.png',dpi=500,bbox_inches = 'tight')
```
